```python
from sklearn.model_selection import train_test_split
x=df.drop('letter',axis=1)
y=df['letter']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
plt.figure(figsize=(15,8))
plt.title("Showing Correlation Between all the Features")
sns.heatmap(x.corr(),annot=True,cmap="viridis")
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled=scaler.fit(x).transform(x)
df_scaled=pd.DataFrame(scaled,columns=df.columns[:-1])
df_scaled.head()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
svc=SVC()
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
```

```python
accuracy_score=accuracy_score(y_test,y_pred)
print(accuracy_score)
```

```python
cm=confusion_matrix(y_test,y_pred)
cm_matrix = pd.DataFrame(data=cm, index=range(1, 27), columns=range(1, 27))
sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='mako')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
