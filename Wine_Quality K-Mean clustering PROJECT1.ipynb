```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r"D:\sofronics\Datasets\wine-quality-white-and-red.csv")
```

```python
df.head()
```

```python
x=df.iloc[:,1:12].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,20):
    kmeans=KMeans (n_clusters=i,init='k-means++',random_state=99)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,20),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=10, init='k-means++', random_state=42)
kmeans.fit(x)
```

```python
# Retrieve labels
labels = kmeans.labels_
```

```python
# Plot the scatter plot
plt.figure(figsize=(16, 12))

for cluster_label in range(10):  # Loop through all 10 clusters
    plt.scatter(
        x[labels == cluster_label, 9],  # alcohol
        x[labels == cluster_label, 5],  # total sulfur dioxide
        alpha=0.5,
        label=f'Cluster {cluster_label}',
    )

plt.xlabel('alcohol', fontsize=18)
plt.ylabel('total sulfur dioxide', fontsize=16)
plt.title('2-Dimensional Scatter Plot with K-Means Clusters', fontsize=20)
plt.legend()
plt.show()
```
