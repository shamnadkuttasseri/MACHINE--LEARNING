```python
# Reading data
data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
# Collecting X and Y
x = data['Head Size(cm^3)'].values
y = data['Brain Weight(grams)'].values
```

```python
x.ndim
```

```python
# Mean X and Y
mean_x = np.mean(x)
mean_y = np.mean(y)
```

```python
# Total number of values
n = len(x)
```

```python
# Using the formula to calculate m c

number = 0
denom = 0
for i in range(n):
    number += (x[i] - mean_x) * (y[i] - mean_y)
    denom += (x[i] - mean_x) **2
m = number/denom
c = mean_y - (m * mean_x)
print("m=",m)

print("c=",c)
```

## y= mx +c

```python
y_new = []

for i in range(n):
    y_new.append((m * x[i]) + c)
```

```python
y_new
```

```python
from matplotlib.figure import Figure

y_new=np.array(y_new)    

fig = plt.figure(figsize = (3, 3))
plt.scatter(x,y,label="Scatter plot")
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.plot(x,y_new,label="Regression line")
plt.legend()
plt.show()
```

# Calculating R^2



```python
# Calculating R^2
ss_t = 0
ss_r = 0
for i in range(n):
    y_pred = c + (m*x[i])
    ss_r += (y[i] - y_pred) ** 2
    ss_t += (y[i] - mean_y) ** 2
r2 = 1-(ss_r/ss_t)
print("R^2:",r2)
```

# Regression analysis using Scikitlearn

```python
x = x.reshape(n,1)
```

```python
x.ndim
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

y_predicted = model.predict(X_test)
```

```python
# calculating r2 score
r2_score = model.score(x,y)

print(r2_score)
```

### Fitting of the model

#### Overfitting, or high variance, 
happens when your hypothesis function hθ(x) tries too hard to fit the training set. The result is that the learned hypothesis function will work great with the training set you initially provided, but will fail to generalize new examples.

#### Underfitting, or high bias: 
the hypothesis function doesn't fit much well the data you have provided in the training stage.
