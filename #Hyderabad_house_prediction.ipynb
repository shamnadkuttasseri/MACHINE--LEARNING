### Removing outliers in continous features

```python
Q1 = data['Price'].quantile(0.25)
Q3 = data['Price'].quantile(0.75)
IQR = Q3 - Q1
filter_price = (data['Price'] >= Q1 - 3.5*IQR) & (data['Price'] <= Q3 + 3.5*IQR)
data = data.loc[filter_price]

### plot boxplot and histogram after removing outliers above and below 3.5 IQR
plt.subplot(1, 2, 1)
_ = sns.boxplot(y='Price', data= data)
plt.subplot(1, 2, 2)
_ = sns.histplot(x='Price', data= data)
plt.tight_layout()
```

```python
# Drop observations greater than some value
data = data[data['Area'] <= 6000]

### plot boxplot and histogram after removing outliers
plt.subplot(1, 2, 1)
_ = sns.boxplot(y='Area', data= data)
plt.subplot(1, 2, 2)
_ = sns.histplot(x='Area', data= data)
plt.tight_layout()
```

### Correlation Analysis

### Feature Scaling
    
####    Apply log to 'Area' for scaling and overcoming the sckewness

```python
data['Area']=np.log1p(data['Area'])
```

#### Apply log to Price for same reason,plus  to avoid negative predicted values

```python
data['Price']=np.log1p(data['Price'])
```

### Selecting higher correlated data

```python
# Calculate the correlation matrix
corr_matrix = data.drop(columns=['Location']).corr()

correlation_threshold = 0.34
high_corr_columns = corr_matrix.loc[(corr_matrix['Price'] > correlation_threshold), 'Price'].index.tolist()

print(f"Columns with correlation above {correlation_threshold} with 'Price':")
print(high_corr_columns)


high_corr_data = data[high_corr_columns]

# Calculate the correlation matrix for the selected columns
corr_matrix_high_corr = high_corr_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data=corr_matrix_high_corr, annot=True, cmap='coolwarm')
plt.title(f'Correlation Heatmap for Columns with Correlation > {correlation_threshold} with Price' )
plt.show()
```

### Model Training

```python
x = high_corr_data.drop(['Price'], axis='columns').values
y = high_corr_data['Price']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

#### Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Modelling
model = LinearRegression()
model.fit(x_train, y_train)

# Prediction
y_pred = model.predict(x_test)

# Calculate R2
r2 = r2_score(y_test, y_pred)

# Calculate adjusted R2
n = len(y_test)
p = x_train.shape[1]  # Number of predictors
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

# Calculate other metrics like Mean Absolute Error (MAE), Mean Squared Error (MSE), and Root Mean Squared Error (RMSE)
mae = np.mean(np.abs(y_pred - y_test))
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Print the results
print(f'R-squared (R2): {r2}')
print(f'Adjusted R-squared: {adjusted_r2}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

### Saving the model

```python
import joblib
joblib.dump(model, 'linear_regression_model.joblib')
```

#### Loading saved model for testing with new data

```python
loaded_model = joblib.load('linear_regression_model.joblib')

Area = 9679000
No_of_Bedrooms = 3
Gymnasium = 0
SwimmingPool = 1
LandscapedGardens = 0
JoggingTrack = 0
RainWaterHarvesting = 1
IndoorGames = 0
ClubHouse = 0
_24X7Security = 1
CarParking = 1
MultipurposeRoom = 1
Children_splayarea = 0

input_data = [Area, No_of_Bedrooms, Gymnasium, SwimmingPool, LandscapedGardens, JoggingTrack, RainWaterHarvesting, IndoorGames, ClubHouse, _24X7Security, CarParking, MultipurposeRoom, Children_splayarea]

output_data = loaded_model.predict([input_data])

print(f'The predicted price is Rs.{round(output_data[0])}/-')
```
