```python
x=df.iloc[:,[1,17]].values
```

```python
from sklearn.metrics.pairwise import euclidean_distances

distance_matrix=euclidean_distances(x)
distance_df=pd.DataFrame(distance_matrix)
distance_df.columns=[f'Distance to point {i+1}' for i in range(distance_df.shape[1])]
distance_df.index=[f'point {i+1}' for i in range(distance_df.shape[1])]

distance_df
```

```python
import scipy.cluster.hierarchy as sch
dendrogram= sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Purchases')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hc=AgglomerativeClustering(n_clusters=3,linkage='ward')
```

```python
y_hc=hc.fit_predict(x)
y_hc
```

```python
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=10,c='red',label='cluster 1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=10,c='blue',label='cluster 2')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=10,c='green',label='cluster 3')

plt.title('Clusters of purchases')
plt.xlabel('Payments')
plt.ylabel('Balances')
plt.legend()
plt.show()
```
